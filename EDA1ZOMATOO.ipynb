```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("zomatoo.csv",encoding="latin-1")
data.head()
```

```python
#only get Integer feature or column 
data.describe()
```

```python
#Find missing nan value
data.isnull().sum()
```

```python
#get only nan column name
[features for features in data.columns if data[features].isnull().sum()>0]
```

```python
#showing nan value using heatmap
import seaborn as sns
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
#walk through xlsx file 
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
#Merginging country code into main table
fdata=pd.merge(data,df_country,on='Country Code' ,how="left")
fdata.head()
```

```python
#to knoow datatype
fdata.dtypes
```

```python
#Accoridng to city County of for eg(Outlets)
fdata.City.value_counts()
```

```python
#Accoridng to country Count of for eg(Outlets)
fdata.Country.value_counts()
```

```python
Country_name=fdata.Country.value_counts().index
Country_name
```

```python
Country_values=fdata.Country.value_counts().values
Country_values
```

```python
plt.pie(Country_values[:3],labels=Country_name[:3],autopct="%2.2f%%")
plt.title("Maximum order or Transaction  according to Country",loc="center")
```

```python
fdata.columns
```

```python
Rating=fdata.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
Rating
#OBSERVATION="ACCORDING TO COUNTRIES CITY THERE RATING OF COUSTMER"
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Aggregate rating",y="Rating Count",data=Rating,palette=['blue','red','orange','yellow','green','darkgreen'])
```

```python
sns.countplot(x="Rating color",data=Rating,palette=['blue','red','orange','yellow','green','darkgreen'])
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='Country',hue="Rating color",data=fdata)
```
